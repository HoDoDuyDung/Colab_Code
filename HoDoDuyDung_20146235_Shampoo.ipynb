```python
# 1 bien bac 3
import pandas as pd
url = 'passenger.csv'
data = pd.read_csv(url)
data.head()

import matplotlib.pyplot as plt
data.hist()
plt.show()
correlation = data.corr()
print(correlation)

import seaborn as sns
sns.heatmap(correlation,annot=True)
sns.pairplot(data)

from sklearn import linear_model
model = linear_model.LinearRegression()
X = pd.DataFrame([data['Miles(X)'],data['Miles(X)']**2,data['Miles(X)']**3]).T
model.fit(X,y=data['Dollars (Y)'])
print('Intercept cua mo hinh :', model.intercept_)
print('Coeffient cua mo hinh :', model.coef_)
print('Do chinh xac cua mo hinh :', model.score(X = pd.DataFrame([data['Miles(X)'],data['Miles(X)']**2,data['Miles(X)']**3]).T, y=data['Dollars (Y)']))

import numpy as np
range = np.arange(0,5500,100) 
X_range = pd.DataFrame([range,range**2,range**3]).T
y_pred = model.predict(X=X_range)
data.plot(kind = 'scatter', x = 'Miles(X)',y='Dollars (Y)',figsize=(9,9),color = 'black')
plt.plot(range,y_pred,color='blue')
```
