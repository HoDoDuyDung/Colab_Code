```python
# 1 bien bac 3
import pandas as pd
url = 'Salary_Data.csv'
data = pd.read_csv(url)
data.head()

import matplotlib.pyplot as plt
data.hist()
plt.show()
correlation = data.corr()
print(correlation)

import seaborn as sns
sns.heatmap(correlation,annot=True)
sns.pairplot(data)

from sklearn import linear_model
model = linear_model.LinearRegression()
X = pd.DataFrame([data['YearsExperience'],data['YearsExperience']**2,data['YearsExperience']**3]).T
model.fit(X,y=data['Salary'])
print('Intercept cua mo hinh :', model.intercept_)
print('Coeffient cua mo hinh :', model.coef_)
print('Do chinh xac cua mo hinh :', model.score(X = pd.DataFrame([data['YearsExperience'],data['YearsExperience']**2,data['YearsExperience']**3]).T, y=data['Salary']))

import numpy as np
range = np.arange(0,12,0.1) 
X_range = pd.DataFrame([range,range**2,range**3]).T
y_pred = model.predict(X=X_range)
data.plot(kind = 'scatter', x = 'YearsExperience',y='Salary',figsize=(9,9),color = 'black')
plt.plot(range,y_pred,color='blue')
```
